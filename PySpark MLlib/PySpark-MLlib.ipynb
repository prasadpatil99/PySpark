### Data Visualisation with Seaborn & Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Create PySpark session & read dataset

```python
from pyspark.sql import SparkSession
from pyspark.sql.types import DoubleType
spark = SparkSession.builder.appName("PysparkML").getOrCreate()
df1 = spark.read.csv('winequality-white.csv',header=True,sep=";",inferSchema=True)
```

Pyspark dataframe does not support much visualisation libraries hence let's continue with pandas dataframe

```python
df_pd = df1.toPandas()
```

Find correlation 

```python
corr = df_pd.corr()
corr.sort_values(["quality"], ascending = False, inplace = True)
print(corr.quality)
```

Convert pandas dataframe into spark dataframe again

```python
df1 = spark.createDataFrame(df_pd)
df = df1.withColumn("label", df1["quality"].cast("double"))        # rename target variable and change its datatype
columns_to_drop = ['quality']                                      # drop old target name 
df = df.drop(*columns_to_drop)     
```

```python
df.show()
```

All the available schemas in dataset

```python
df.printSchema()
```

Vector Assembler concatenates multiple feature into single vector to further preprocess

```python
from pyspark.ml.feature import VectorAssembler, StandardScaler
feature_assembler=VectorAssembler(
    inputCols=["fixed acidity","volatile acidity","citric acid","residual sugar",
                                            "chlorides","free sulfur dioxide","total sulfur dioxide","density",
                                            "pH","sulphates","alcohol"], outputCol="Independent_Feat")

output=feature_assembler.transform(df)
```

Scale all the features 

```python
sscaler = StandardScaler(
    withMean=True, withStd=True, inputCol="Independent_Feat", outputCol="features"
)
scaled = sscaler.fit(output).transform(output)
scaled.select("Independent_Feat","features").show(5)
```

Split Dataset into train and test dataset

```python
train_data,test_data=scaled.randomSplit([0.75,0.25], seed = 42)
```

### Training model via LogisticRegression
LogisticRegression is actually a classification algorithm which is somewhat similar to linear regression just logistic provides differing probability from other classes containing continous quantity while linear regression predicts raw unknown value of 'x' 

```python
from pyspark.ml.classification import LogisticRegression
lr = LogisticRegression(featuresCol = 'features', labelCol = 'label')
lr_model = lr.fit(train_data)

predictions = lr_model.transform(test_data)
```

Let's evaluate the results 

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
evaluator = BinaryClassificationEvaluator(rawPredictionCol='prediction', labelCol='label', metricName='areaUnderROC')
print("The area under ROC for test set is {}".format(evaluator.evaluate(predictions)))
```

```python
predictions.select("label","prediction").show(10)
```

Results are already good but though let's try tuning different parameters

## Hyperparameter Tuning
Hyperparameter tuning or optimization is a technique of choosing optimal hyperparameters for algorithm <br>
which helps in increasing performance of model. <br>
Hyperparameters used here :<br>
elasticNetParam - *for L1 or L2 penalty*<br>
fitIntercept - *whether to add bias* <br>
maxIter - *max no of iterations*<br>
regParam - *regularization parameter*<br>


```python
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
import numpy as np

paramGrid = ParamGridBuilder()\
    .addGrid(lr.elasticNetParam,[0.0, 1.0])\
    .addGrid(lr.fitIntercept,[False, True])\
    .addGrid(lr.maxIter,[10, 50, 100])\
    .addGrid(lr.regParam,[0.01, 0.5, 2.0]) \
    .build()

from pyspark.ml.evaluation import BinaryClassificationEvaluator
crossval_lr = CrossValidator(estimator=lr,
                          estimatorParamMaps=paramGrid,
                          evaluator=BinaryClassificationEvaluator(),
                          numFolds= 5)  

cvmodel_lr = crossval_lr.fit(train_data)
```

```python
predict_train=cvmodel_lr.transform(train_data)
predict_test=cvmodel_lr.transform(test_data)
```
